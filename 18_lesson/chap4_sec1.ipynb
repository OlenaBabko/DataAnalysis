```python
import matplotlib.pyplot as plt
import numpy as np

A=np.loadtxt('data.txt',delimiter=None)
x=A[:,0]
y=A[:,1]
plt.plot(x,y,lw=2,color='blue')
plt.xlabel('Gate Voltage (V)')
plt.ylabel('Resistance (k$\Omega$)')
```

```python
import matplotlib.pyplot as plt
import numpy as np

A=np.loadtxt('data1.txt',delimiter=None,skiprows=1)
x=A[:,0]
y=A[:,1]
plt.plot(x,y,lw=2,color='blue')
plt.xlabel('Gate Voltage (V)')
plt.ylabel('Resistance (k$\Omega$)')
```
